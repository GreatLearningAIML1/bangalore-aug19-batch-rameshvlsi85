# 1) Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. 

```python
import numpy as np
import pandas as pd
```

```python
llg=pd.read_csv("Laliga.csv")
llg.replace('-','0',inplace= True)

print(llg)
```

```python
#the column name for the laliga is Unnamed
print(llg.columns)

#Create new dataframe new_llg, with [0] row as new column name and remove the row with index[0] 

new_llg=llg.rename(columns=llg.iloc[0]).drop(llg.index[0])
print('\n\n\n\n',new_llg)
```

# 2) Print all the teams which have started playing between 1930-1980. 

```python
#split the column 'Debut' using delimiter hypen '-' and 
#store the year values in new Series sy(start year) ey(end year)

split_debut=new_llg['Debut'].str.split('-',1)
sy =split_debut.str.get(0)
ey =split_debut.str.get(1)


#fetching the index from sy, store value ey = sy +1 and append to the list end_yr

for i in sy.index:
    ey.at[i] = int(sy.at[i]) +1
        
# creating new columns 'Debut_Start' and 'Debut_End' in new_llg
new_llg['Debut_Start']=sy.astype('int')
new_llg['Debut_End']=ey.astype('int')
```

```python
#Teams which have started playing between 1930-1980 

debut_new = new_llg.loc[:,['Team','Debut']][(new_llg['Debut_Start']>= 1930) & (new_llg['Debut_End']<=1980)]


#remove the index from the dataframe

debut_new_value= debut_new.to_string(index=False)
print('Teams which have started playing between 1930-1980 are as follows: \n\n',debut_new_value)
    
```

# 3) Print the list of teams which came Top 5 in terms of points 

```python
# Sort the dataframe based on the values in column 'Points'
# Then fetch the top five teams from sorted dataframe

new_llg['Points'] = new_llg['Points'].astype('int')
new_llg.sort_values(by='Points', ascending=False)
top5 = new_llg['Team'].head(5)

top5_value = top5.to_string(index=False)
print('Top five teams based on points are: \n',top5_value)
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference

# Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count():
    new_llg['Goal_diff_count'] = new_llg['GoalsFor'].astype('int') - new_llg['GoalsAgainst'].astype('int')
    return new_llg.loc[:, ['Team', 'Goal_diff_count']]

goal_diff = Goal_diff_count()

#print(goal_diff)
#Find the team which has maxmum and minimum goal difference
g_max = goal_diff['Team'][goal_diff.Goal_diff_count == goal_diff.Goal_diff_count.max()]
g_min = goal_diff['Team'][goal_diff.Goal_diff_count == goal_diff.Goal_diff_count.min()]

# Team name without index:
g_max_value= g_max.to_string(index=False)
g_min_value= g_min.to_string(index=False)

print('\nThe team which has maximum goal difference is: ',g_max_value)
print('\nThe team which has minimum goal difference is: ', g_min_value) 
```

# 5) Create a new column with name “Winning Percent” and append it to the data set  
# Percentage of Winning = (GamesWon / GamesPlayed)*100
# If there are any numerical error, replace it with 0%
# Print the top 5 teams which has the highest Winning percentage

```python
#created new column 
new_llg["Winning Percent"] = (new_llg['GamesWon'].astype('float')/new_llg['GamesPlayed'].astype('float'))*100

#replaces NaN with 0
new_llg.fillna(value=0, inplace=True)

#Sort by "Winning Percent" and print top 5 Teams
sort_by_per = new_llg.sort_values(by="Winning Percent", ascending=False)
top5_win = sort_by_per.loc[:,['Team' ,'Winning Percent']].head(5)
top5_win_value = top5_win.to_string(index=False)
print('The top 5 teams which has the highest Winning percentage are: \n\n',top5_win_value)
```

# Eg: Best Position Points

```python
#change the type of 'BestPosition' to int and sort the order by 'BestPosition' value

new_llg['BestPosition']=new_llg['BestPosition'].astype('int')
sort_by_BP = new_llg.sort_values(['BestPosition'],ascending=True)

#grp_by_BP = sort_by_BP.groupby('BestPosition').sum()[['Points']] –» This returns a DataFrame object.

#grp_by_BP = sort_by_BP.groupby('BestPosition').sum().Points –» This returns a Series object.
#convert series to dataframe by using df.to_frame()
#print(grp_by_BP.to_frame())

grp_by_BP = sort_by_BP.groupby('BestPosition').sum()[['Points']] 
print('The sum of points for all positions is as below: \n\n',grp_by_BP)

grp_by_BPT = sort_by_BP.groupby(['BestPosition','Team']).sum()[['Points']] 
print('\n\n\nGroup of teams based on their “Best position” is: \n\n',grp_by_BPT)
```
