### Plot the attrition column (Since it is a categorical variable, chose the plot accordingly)

```python
Yes=0
No=0
for x in hr['Attrition']:
   if(x=='Yes'):
    Yes=Yes+1
   elif(x=='No'):
    No=No+1
print('The count of Yes is:',Yes,'\nThe count of No is:',No)
df = pd.DataFrame({'Attrition':['Yes', 'No'], 'count':[Yes, No]})
ax = df.plot.bar(x='Attrition', y='count',rot=0)
```
